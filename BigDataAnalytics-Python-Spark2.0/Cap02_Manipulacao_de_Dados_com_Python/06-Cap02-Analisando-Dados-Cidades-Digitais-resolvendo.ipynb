```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
del cidadesDigitais['IBGE']
```

```python
# Aplicando Group By
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
cidadesUF_1.groups
```
